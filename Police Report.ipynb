#### Bar Chart

```python
#Select y Variable from stopsByRace
y = pd.Series(stopsByRace.stops)

# Select x Variable
x = pd.Series(stopsByRace.race)

ax = y.plot(kind='bar', figsize=(15, 10), color = "red", )

ax.set_xlabel('Race')
ax.set_ylabel('Number of Stops')

ax.set_title('Number of Police Stops by Race in San Diego')

ax.set_xticklabels(x)

# add labels to each bar in the chart

def add_value_labels(ax, spacing=5):
    
    #For each bar: Place a label
    for rect in ax.patches:
        # Get X and Y placement of label from rect
        y_value =rect.get_height()
        x_value =rect.get_x() + rect.get_width() / 2
        
        #Number of Points Between bar and Label. Change to your liking
        space = spacing
        #Vertical aligment for positive values
        va = 'bottom'
        
        #If value of bar is negative: Place label below bar
        if y_value < 0:
            #Invert space to place label below
            space *=  -1
            #Vertically align Label at top
            va = 'top'
            
        #Use Y Value as label and format number with one decimal place
        label = "{:.1f}".format(y_value)
        
        #Create annotation
        ax.annotate(                    
        label,                         #Use label as 'Label'
        (x_value,y_value),             #Place label at the enf of the bar
        xytext=(0, space),             #Vertically shift Label by 'space'
        textcoords="offset points",    # Interpret 'xytext' as offset in points
        ha='center',                   # Horizontally center label
        va=va)                         # Vertically align label differently for
                                       #positive and negative values
            
            
add_value_labels(ax)
        


```

#### Box Plot

```python
# To see min and max values, quartiles and outliers

import seaborn as sns

#set the style
sns.set_theme(style="whitegrid")

#set the size
plt.figure(figsize =(15,10))

#plot the variable stop duration separated by race from datset df_reduced
ax = sns.boxplot(x="race", y="stop_duration", data=df_reduced)

ax.set_xlabel('Race')
ax.set_xlabel('Stop Duration')
ax.set_title('Boxplot of Race vs. Stop Duration')

```

Because of the outliers we can not see the distribuition well. 

```python
plt.figure(figsize =(500,500))

# displots (histogram)
#kde= false dissables the kernel density line plotted over the histogram
ax =sns.displot(df_reduced.stop_duration,height = 10, aspect =2,  kde=False)

ax.set(title = 'Distribuition of stop_duration')
ax.set_ylabel('Count')
```

The previous graph look as they do because of the outliers. The variables are skewed to the left and in a range of 0 to 1400. while a value of 1400 minutes certaily seems an outlier, but we can consult the Empirical Rule to be sure.

The Empirical Rule saya the following

- About 68% of our data lies within one standard deviation of the mean value
- About 95% of our data lies within two stardard deviations of the mean value
- About 99.7% of our data lies within three standard deviations of the mean value

https://www.statistichowto.com/empirical-rule-2/

So, we can calculate the mean and standard deviation of our stop_duration data using numpy packages. Then, we can select all rows in our dataset where stop_duration is within 2 stardard deviations of the mean and still be left with 95% of our data.

```python
#mean

mean = np.mean(df_reduced.stop_duration)

# standard deviation 
sd = np.std(df_reduced.stop_duration)

# select all rows in the df_reduced dataset where stop duration is less than our mean plus 2 SDs and more than our mean minus 2 SDs
df_duration_trimmed = df_reduced.loc[(df_reduced['stop_duration'] < mean + sd*2) & (df_reduced['stop_duration'] > mean -sd*2)]
```

Visualizing the new distribuition

```python
plt.figure(figsize =(10,10))

ax = sns.displot(df_duration_trimmed.stop_duration,height = 10, aspect = 2, kde=False)

ax.set(title = 'Distribuition of stop_duration')
ax.set_ylabel('Count')
```

Boxplot with new distribuition

```python
plt.figure(figsize =(15,10))

ax = sns.boxplot(x='race', y='stop_duration', data=df_duration_trimmed)

ax.set_xlabel('Race')
ax.set_ylabel('Stop Duration')
ax.set(title =' Boxplot of Race vs. Stop Duration (without Outliers)')
```

### Featuring Engineering

Now, we will turn our continuous text variables to binary variables we can use in more in-depth analysis. 

#### Transformation with skewed continuous variable

Let's look at something else we can do to make stop_duration a bit more managable, transformation. Now, the type of transformation used depends on the type of skewness the variable exhibits.

```python
plt.figure(figsize =(10,10))

ax = sns.distplot(df_reduced.stop_duration, kde=False)

ax.set(title = 'Distribution of stop_duration')
ax.set_ylabel('Count')
```

Since we are skewed to the left we can use logarithm or Inverse.In that case, we can use a logarithmic transformation to smooth out the distribution of the variable and bring it closer to normal. A variable having a close to normal distribution is crucial for many machine learning algorithms to function correctly.  Transformation takes the natural log of that point to create a new distribuition that has been iterated through with that mathematical operation.

```python
# apply numpy function log() to automatically take the natural log of each value in stop_duration in the dataframe df_reduced
duration_transformed = np.log(df_reduced.stop_duration)

plt.figure(figsize =(10,10))

ax = sns.distplot(duration_transformed, kde=False)

ax.set(title = 'Distribution of stop_duration (with log transformation')
ax.set_ylabel('Count')
```

This distribuition is closer to a normal what is ideal for machine learning

#### Binarization

Binarization is a method of feature engineering that turns categorical or continuous variables into 0s and 1s. Often, binarization is used to allow the use of classification algorithms.

Firts, we will look at race variables and create a black variable. 1 if the motorist is black 0 if the motorist is any other race. For this we need to look the distribuition of stops across races.

```python
df_reduced.race.value_counts()
```

```python
# Now, we wil create a binary field

condition = (df_reduced['race'] == 'Black/African American')

# Create a new variable black and if condition is true then 1, if condition is false, 0.
df_reduced['black'] = np.where(condition,1,0)
```

```python
df_reduced.black.value_counts()
```

With the above we see that 285,137 of the motorists stopped are not black, and 71,069 are black.

```python
# Now, we will create a variable based on gender by selecting those who are labeled as 'male' or 'transgeder men/boy'
condition = (df_reduced['gender'] == 'Male') | (df_reduced['gender'] == 'Transgender man/boy')

# create new male variable
df_reduced['male'] = np.where(condition, 1,0)
```

```python
df_reduced.male.value_counts()
```
